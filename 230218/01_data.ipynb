```python
### 데이터프레임의 결합
# 1. 단순한 행, 열 결합
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    's1' : [1, 2, 3, 4],
    's2' : [5, 6, 7, 8]
})
df
```

```python
df2 = pd.DataFrame([1,2])
df2
```

```python
### concat()
# 1. 데이터 프레임을 결합하는 함수
# 2. 매개변수 axis -> 행을 추가할지 열을 추가할지 지정
# 3. 단순하게 데이터프레임을 결합하는 함수
# 4. 매개변수 ignore_index -> 기본값은 false 이며 인덱스의 값을 그대로 유지할지 지정하는 매개변수
```

```python
## 단순하게 행을 추가
pd.concat([df,df2], axis = 0) # 단순하게 결합한 결과를 볼 수 있음
```

```python
## 단순하게 행을 추가 하고 인덱스를 재 지정 : 위의 인덱스가 중복된(0, 1dl..) 것을 삭제하고 재 지정하는 것

## case1
# pd.concat([df,df2], axis=0).reset_index(drop=True)

## case2
pd.concat([df,df2], axis =0, ignore_index = True)
```

```python
## 열을 추가(열로 결합)
pd.concat([df,df2], axis = 1)
```

```python
df3 = pd.DataFrame({
    's1' : [1,2]
})
df3
```

```python
pd.concat([df,df3],axis=0) # df3 에는 s1 이 있으므로 s1 열에 df3 값이 행으로 추가됨
```

```python
### merge()
# 1. 데이터프레임을 결합하는 함수
# 2. 특정 조건에 맞춰서 (기본적으로) 열을 추가
# 3. 매개변수 on -> 조건(을 지정)
# 4. 매개변수 how -> 데이터프레임에 기준점 (left, right, inner, outer)
#    : 어떻게 합쳐줄 것이냐 방법을 정하는 것..inner(교집합), outer(합집합), left(왼쪽의 데이터프레임..)
```

```python
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "05190", "LG 화학", 896000],
    ["전기전자", "000660", "SK하이닉스", 101500]
]
columns = ["업종", "종목코드", "종목명", "현재가"]

df = pd.DataFrame(data=data, columns=columns)
df
```

```python
data = [
    ["은행", 2.92],
    ["보험", 0.37],
    ["화학", 0.06],
    ["전기전자", -2.43]
]
columns = ["업종", "등락률"]

df2 = pd.DataFrame(data = data, columns=columns)
df2
```

```python
## 데이터베이스에서 컬럼이 많을 수록 부하가 많이 걸림(인덱스는 별로 영향 x). 그래서 컬럼을 적게 만들어 필요할 때마다 결합을 시킴
```

```python
pd.merge(left = df, right = df2, on="업종", how="left") # concat 은 그냥 막 합친 것이고..이건 조건을 지정해서 합쳤기에 보험이 없음
                                                        # left 에 df 가 들어갔기에 df 를 기준으로 합친 것이라 보험이 없음
                                                        # 즉 왼쪽을 살리고 왼쪽에 없는 오른쪽은 것은 지우라는 것
```

```python
pd.merge(left = df, right = df2, on="업종", how="right") # 오른쪽을 기준으로 살렸기에 보험도 표시됨. right 에 df2 가 들어갔기에 df2 를 
                                                         # 기준으로 합친다는 것
                                                         # 즉 오른쪽을 살리고 오른쪽에 없는 왼쪽은 것은 지우라는 것
```

```python
df3 = pd.DataFrame({
    "업종" : ["서비스업"],
    "종목코드" : ["035720"],
    "종목명" : ["카카오"],
    "현재가" : [121500]
})
df3
```

```python
df4 = pd.concat([df,df3], axis = 0, ignore_index = True)
df4
```

```python
pd.merge(df4, df2, on='업종', how='left')   # left, right 는 처음나오는 것이라 안적어도 되긴 함
```

```python
pd.merge(df4, df2, on='업종', how='right')   # left, right 는 처음나오는 것이라 안적어도 되긴 함
                                             # 위와의 차이를 잘 보라
```

```python
pd.merge(df4, df2, on='업종', how='inner') # 교집합
```

```python
pd.merge(df4, df2, on='업종', how='outer') # 합집합
```
