```python
### 예제 문제
1. csv 폴더에 있는 sales records.csv 파일 로드
2. "Order ID", "Total Revnue", "Total Cost", "Total profit" 칼럼 삭제
3. 컬럼의 이름을 변경 ['대륙', '국가', '타입', '채널','우선순위','주문일자','발송일자','판매단위','단가','원가'] 로 변경
4. 대륙, 국가를 기준으로 오름차순 정렬하고
5. 인덱스를 초기화(기존의 인덱스는 삭제)
```

```python
import pandas as pd ## 새로운 file 을 만들었으므로 판다스 로드가 필요함
```

```python
df = pd.read_csv("../csv/Sales Records.csv")
```

```python
df
```

```python
df = df.drop(["Order ID","Total Revenue","Total Cost","Total Profit"],axis=1) 
  ## df = df.drop 대신 df.drop에 inplace= True 를 사용해도 됨
  ## 각 열의 이름을 각각 리스트로 하는 것이 아니라 지우고자 하는 전체 열을 리스트로 묶어야 함..
  ## 매개변수가 하나이므로 대괄호를 한개 즉 하나의 리스트로 표시 해야한다는 것......공부 더 해야 함...
```

```python
df
```

```python
df.columns= ['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
df
## df.head() 만 하면 칼럼이 변경된 것만 확인할 수 있음
```

```python
df = df.sort_values(['대륙','국가']) ## 이거 실행해서 아무것도 안나와야 sort 결과가 적용된 것임. 적는 순서대로 오름차순 정렬이 됨
                                    ## df.sort_values(['대륙','국가']) 만하면 sort 하면 어떻게 되는지만 보여주는 것임
                                    ## df = df.sort 하지 않고..df.sort 안에 replace=True 로 변경해줘도 됨
                                    ## values 에 매개변수가 한개만 들어가므로 리스트 하나 라는 것...                                    
```

```python
df
```

```python
## 마지막 숙제 인덱스 초기와 방법(이것을 못 풀었음...시간이 부족???ㅠㅠ)
# 방법 1
df.reset_index(drop=True) # df.reset_index() 만 하면 df 를 안바꾸고 실행될 때의 내용을 보여주기만 함
# 방법 2
df = df.reset_index().drop(['index'],axis=1) 
   # df.reset_index().drop(['index'],axis=1) 만 하면 df 를 안바꾸고 실행될 때의 내용을 보여주기만 함
```

```python
df.head(3)
```

```python
## 총단가, 총원가 칼럼 추가하여 구하고 
## 총 이윤 칼럼 추가하여 구해봐라
```

```python
df['총단가'] = df['판매단위']*df['단가'] # df(["판매단위"]*["단가"]) 하면 안됨. df 안의 판매단위 와 단가의 각각 칼럼을 곱해주는 것이라...
df['총원가'] = df['판매단위']*df['원가']
df['총이윤'] = df['총단가']-df['총원가'] # df['총이윤'] = df['판매단위]*(df['단가']-df['원가']) 도 동일함
```

```python
df # df.head(3) 하면 위의 세개 행, df.tail(3) 하면 아래의 3개행을 보여 줌
```

```python
### loc[]  : location 의 약자이고 특정한 위치에 data 출력을 위해 사용
    ## - 특정 행이나 특정 열을 추출할 때 사용
    ## loc[행의 위치, 열의 위치]  열의 위치를 안적어주면 모든 열이 표시

### iloc[] : 위치기반의 특정한 위치에 data 출력을 위해 사용
    ## 특정 행이나 특정 열을 추출할 때 사용(목적은 위와 동일)
    ## iloc[행의 위치 인덱스 값, 열의 위치 인덱스 값]
```

```python
df.drop([0],axis=0, inplace=True)
```

```python
# df.loc[0] # 이렇게 하면 애러가 나옴 왜냐하면 index 0 이라는 값이 없으므로
df.iloc[0] # 이렇게 하면 나옴. 왜냐하면 여기서의 0은 위치 값이므로 첫번째의 위치가 나옴
```

```python
df.drop([2],axis=0, inplace=True) # 아래 loc 와 iloc 의 결과 값을 잘 비교해 보고 이해해라
```

```python
df.loc[:,'국가'] # 인덱스는 전체 컬럼은 국가만 뽑으라 는 것임..인덱스 아리에 20:22 이면 20~22 인덱스의 국가만 가져 옴 
                 # '국가':'단가' 하면 국가~단가까지
                 # 똑같이 iloc 에서 출력하려면 어떻게 쓸지 고민해봐라..
```

```python
### 그룹화 : 만들은 df 의 각 컬럼별로 그룹화를 할 수 있음
## groupby(그룹화 기준이 되는 컬럼)
```

```python
# df.groupby("국가").sum() # 그룹화 한 것의 숫자 형태를 모두 더해서 보여줌
df.groupby("국가").sum()['총이윤'] # 위에 모든 것을 보여준 것에서 '총이윤'만 선택하여 보여줌
                                  # df.groupby("국가").sum()[['총이윤']] 이렇게 하면 데이터 프레임 형태로 출력
```

```python
## 가장 많은 이윤을 내는 대륙은 어디인가? -> 그룹화 하고 내림차순 또는 오름차순 하여 상위값 또는 확인해야함

# df.groupby('대륙').sum()['총이윤']
test = df.groupby('대륙').sum()[['총이윤']] # 위의 것을 test 라는 변수에 저장해줘야 함, 총이윤에 [[]] 인 것은 df 형태로 만든 것이고..
```

```python
## 내림차순 정리하고 해드값 하나 확인
test.sort_values(['총이윤'], ascending=False).head(3) # .head() 없으면 다 보여줌
   # 위에 총이윤을 [[]] 로 데이터 프레임 형태로 했으므로 총이윤을 칼럼으로 확인할 수 있음
   # 만약 위에서 총이윤을 [] 만 하면 스리즈 형태가 되므로 스리지는 칼럼이 없기에 여기서 실행하면 에러가 발생함
```

```python
## 국가별 총 이윤이 높은 상위 10개 국가를 출력 : 내림차순 정리해서 iloc 를 써서 해도 되고 head(10) 해도 됨

test2 = df.groupby('국가').sum()[['총이윤']]
## 총 이윤을 기준으로 내림차순 정리
group_df = test2.sort_values(['총이윤'],ascending=False)
## 상위 10개
top10 = group_df.head(10)
## 하위 10개
bottom10 = group_df.tail(10)
```
