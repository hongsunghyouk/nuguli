```python
## 라이브러리 로드
import pandas as pd
```

```python
## 코로나 데이터를 로드
## 상위폴더로 이동해서 csv 폴더로 이동해서 corona.csv 파일 지정
corona = pd.read_csv("../csv/corona.csv")
corona.head()
```

```python
## 컬럼을 삭제
## drop() 함수를 이용하여 컬럼을 삭제
## axis 라는 매개변수의 값을 기준으로 행, 열을 지정해야 함
## replace 매개변수의 값을 기준으로 기준이 되는 데이터를 변경을 할지 지정

corona.drop(["Unnamed: 0"], axis=1) # 결과 값이 아래와 같이 나오는 것은 기존의 data 는 변경이 없다는 것
```

```python
corona.drop(["Unnamed: 0"], axis=1, inplace=True) # inplace = False(inplace 의 기본값은 False 임) 로 하면 기존 index 를 유지하고 새로하나 
                                                  # 더 생기는 것이고..inplace=False 는 기존값을 바꾸지 마라는 것
                                                  # inplace 앞까지는 확인하는 것이고 그 확인을 data 에 적용하라는 명령이 inplace 라고 보면 됨
                                                  # inplace 의 기본값이 false 이므로 안적으면 그냥 false 이므로 적용이 안됨
# corona = corona.drop(["Unnamed: 0"], axis=1) # 위의 것과 동일한 결과를 가져오는 것. inplace 를 모를때 사용 가능
                                             # 두개다 실행하면 에러가 발생하므로 주석처리하여 한개만 실행되어야 함
                                             # 값이 변경되었으므로 실행하면 아무것도 안나옴
```

```python
corona.head() # 제일 앞의 Unnamed 가 있던 컬럼이 없어진 것을 볼 수 있음
```

```python
## 컬럼의 목록을 출력하는 방법
corona.columns
```

```python
## 컬럼의 이름을 변경하는 방법, 반드시 갯수를 동일하게 만들어 줘야 함. 아래만 실행시키면 값이 바뀌었기 때문에 아무것도 안나오는 것이 정상
corona.columns = ["등록일시", "총 사망자", "총 확진자", "게시글 번호",
                  "기준일", "기준시간", "수정일시", "누적 의심자", 
                  "누적 확진율"]
```

```python
corona.head() # 위의 실행에 대한 변경된 결과를 확인.()에 숫자 지정안했으니 기본 값이 5개가 나옴
```

```python
# 게시글 번호, 기준일, 기준시간 컬럼 삭제
corona.drop(["게시글 번호", "기준일", "기준시간"], axis=1) # 실행 했을 때 값이 출력되므로 아직 원본 data 는 바뀌지 않음
```

```python
# 게시글 번호, 기준일, 기준시간 컬럼 삭제 하고 기존의 값을 변경
corona.drop(["게시글 번호", "기준일", "기준시간"], axis=1, inplace=True) # 실행 했을 때 값이 출력안되므로 원본 data 가 바뀌었음
# corona = corona.drop(["게시글 번호", "기준일", "기준시간"], axis=1) # inplace=Ture 와 동일한 결과 corona = corona.dorp..으로 하여 치환
```

```python
### 데이터 프레임의 정렬을 변경
### 등록일시를 기준으로 오름차순 정렬
### sort_values([기준이 되는 컬럼명])

corona.sort_values(["등록일시"]) # 등록일시 순서가 바뀐 것이 확인됨
```

```python
### 데이터 프레임의 정렬을 변경
### 등록일시를 기준으로 오름차순 정렬
### sort_values([기준이 되는 컬럼명])
### ascending 매개변수의 값이 True(기본값) 면 오름차순, False 면 내림차순 정렬

corona.sort_values(["등록일시"], inplace=True) # 실행결과가 나오지 않으므로 정상 작동되었고 오름차순 기준으로 변경(index 순서 변경됨)되어 저장됨
```

```python
## 인덱스 값을 초기화 
## reset_index() : 기존의 index 를 하나의 새로운 컬럼으로 만들어 보여줌
corona.reset_index()
```

```python
## 인덱스 값을 초기화 
## reset_index() : 기존의 index 를 하나의 새로운 컬럼으로 만들어 보여줌
corona.reset_index().drop(["index"], axis=1) # 새로운 것이 생겼으므로 기존의 index 를 삭제하라는 것. 
                                             # inplace 가 없으므로 보여주시만 하는 것. 아래 결과값이 출력되었으므로 실제 data 는 안바뀐 것임
# corona.reset_index(drop=True) #위의 것과 동일한 내용인데 axis 를 할용하지 않음
```

```python
## 인덱스 값을 초기화 
## reset_index() : 기존의 index 를 하나의 새로운 컬럼으로 만들어 보여줌
# corona.reset_index().drop(["index"], axis=1, inplace=True) # 새로운 것이 생겼으므로 기존의 index 를 삭제하라는 것 이고 위와 다르게 실제 data 를 바꿈
corona.reset_index(drop=True, inplace=True) # 위의 것과 동일한 내용인데 axis 를 할용하지 않음
```

```python
## 데이터 간의 차이를 출력하는 함수
## diff(n) : n은 인덱스의 이동 횟수(n은 기본값이 1). 값들의 차이를 구하는 것. 첫번째가 Nan 라고 나오는 것은 전 data(비교 data가 없으므로)
corona["총 확진자"].diff()
```

```python
## 상기의 컬럼을 추가하는 것
corona["일일 확진자"] = corona["총 확진자"].diff() # 딕셔너리에 key 와 value 값을 추가하는 것과 동일 방법
```

```python
corona.head()
```

```python
# 일일 사망자 구하고 칼럼 추가하는 방법
corona["일일 사망자"] = corona["총 사망자"].diff()
```

```python
corona.head()
```

```python
## 인덱스를 이동
## shift(n) n 만큼 인덱스를 이동하는 것

corona["총 사망자"].shift(1) # 기존에 있는 값에서 바로 위에 있는 값을 출력. diff() 는 차이이고 이것을 바로 위의 data를 보여줌
                            # 아래의 data 와 비교하면 한줄식 이동해 있음
```

```python
corona["총 사망자"]
```

```python
corona["총 사망자"] - corona["총 사망자"].shift(1) # diff() 와 동일한 결과를 가져온다. shift 는 그 전의 data 를 보여줄 필요가 있을 때 사용
```

```python
## 결측치를 다른 값으로 대체
## fillna() 채운다(fill) 결측치(na) 의 뜻
## fillna(n) : 결측치를 n 으로 채워준다

## 일일 확진자, 일일 사망자의 결측치를 0 으로 대체
corona[["일일 확진자", "일일 사망자"]].fillna(0) # 하기 출력이 된다는 것은 아직 값이 바뀌지 않음
```

```python
## 결측치를 다른 값으로 대체
## fillna() 채운다(fill) 결측치(na) 의 뜻
## fillna(n) : 결측치를 n 으로 채워준다

## 일일 확진자, 일일 사망자의 결측치를 0 으로 대체
corona[["일일 확진자", "일일 사망자"]]= corona[["일일 확진자", "일일 사망자"]].fillna(0) # 출력안되므로 실제 data 를 바꾼 것
```

```python
corona.head()
```

```python
## fillna() 에는 method 매개변수가 두개 있음 -> ffill(바로 앞의 값으로 대체), bfill(바로 뒤의 값으로 대체)
```

```python
corona["누적 의심자"] = corona["누적 의심자"].fillna(method="ffill")
```

```python
## 라인 그래프 출력

## 도화지에 그려질 x, y 축의 수치를 지정
plt.plot([1,2,3,4]) # 인자값을 하나만 지정하면 y 축 , []로 한개만 적용하였으므로 인자가 하나라는 의미
                    # 인자가 두개면 앞의 것이 x 축, 뒤의 것이 y 축

## 그래프 출력
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [3, 6, 2, 7]
y2 =  [1, 10, 3, 1]

## 선의 색을 바꿔는 것
plt.plot(x, y1, "r")

## 선의 종류를 바꾸는 것
plt.plot(x, y2, "--") ## "--r" 로 하면 선(--)색이 바뀜
# plt.show() # 이것이 주석 처리되지 않으면 같은 코드 내에서 show 가 두번이 나오기 때문에 밑의 그래프는 안나오고 공백이 됨

## x 축의 이름
plt.xlabel("Class")
## y 축의 이름
plt.ylabel("Count")
## title
plt.title("Graph")
## 범례 추가
plt.legend(["A","B"])
plt.show()
```

```python
## subplt(행의 수, 열의 수, 위치) : 그래프를 여러개 같이 보여줌, 행,열, 위치 지정하여 보여 줌

## 그래프 2개를 그린다는 가정

## subplot 을 이용하여 위치를 지정
plt.subplot(2,1,1) # 상, 하로 나누고 1번(상)의 위치를 선택한 것
plt.plot(x,y1) # 1번의 위치에 그릴 그림
plt.subplot(2,1,2) # 상하로 나누고 2번(하)의 위치를 선택한 것
plt.plot(x,y2) # 위의 위치에 그릴 그림

plt.show()
```
